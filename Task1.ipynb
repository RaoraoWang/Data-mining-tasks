```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
data_path = 'D:\\data_mining_wx\\task1\\data.csv'
df = pd.read_csv(data_path)
###查看数据信息
df.info()
###4754个样本，90列
```

```python
delete_cols = []
for col_name in df.columns:
    if df[col_name].dtype == 'object':
        print col_name
        delete_cols.append(col_name)
###待删除的特征
print delete_cols
```

```python
df = df.drop(delete_cols,axis=1)
###删除unnamed列
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
###特征类型转换，floeat
df = df[df.columns] = df[df.columns].astype(float)
df.info()
```

```python
###删除缺失值
df = df.dropna()
df.info()
```

```python
##构建lable
y = df['status']
###构建feature
X = df.drop(['status'],axis=1)
###切割训练测试集合
x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=2018)
###xgb分类器
clf = xgb.XGBClassifier()
clf = clf.fit(x_train, y_train)
predict = clf.predict(x_test)
###输出准确率
accuracy = accuracy_score(y_test, predict)
print accuracy
```
